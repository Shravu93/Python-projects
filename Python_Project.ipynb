## Exploratory Data **Analysis**

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender', data = df)

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'])['Amount'].sum().reset_index()
```

```python
sales_gen
```

```python
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

# from Above graphs we can see that most of the buyers are females

# Age

```python
df.columns
```

```python
ax = sns.countplot( data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# Total Amount vs Age Group

sales_age = df.groupby(['Age Group'])['Amount'].sum().reset_index()
```

```python
sales_age
```

```python
sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)
```

# **From this graphs we can see that most of the buyers are of age group between 26-35 years**

# Sates

```python
df.columns
```

```python
sales_state = df.groupby(['State'])['Orders'].sum().reset_index().head(10)
```

```python
sales_state
```

```python
sns.set(rc={'figure.figsize': (15,5)})
```

```python
sns.barplot(data =sales_state, x = "State", y = "Orders" )
```

```python
sales_state = df.groupby(['State'])['Amount'].sum().reset_index().head(10)
```

```python
sales_state
```

```python
sns.set(rc={'figure.figsize': (15,5)})
```

```python
sns.barplot(data =sales_state, x = "State", y = "Amount" )
```

# From above graphs we can see that most of the orders and total sales amount are from Karnataka.

# Marital Status

```python
sns.set(rc ={'figure.figsize':(5,5)})
```

```python
ax = sns.countplot( x = "Marital_Status", data = df)

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_age = df.groupby(["Marital_Status",'Gender'])["Amount"].sum().reset_index()
```

```python
sales_age
```

```python
sns.barplot(x = "Marital_Status", y = 'Amount', hue = 'Gender', data = sales_age)

```

# From above graphs we can see that most of the buyers are married (female) & they have high purchasing power.

# Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
```

```python
ax = sns.countplot(x = 'Occupation', data = df )

for bars in ax.containers:
  ax.bar_label(bars)
```

# Product categories

```python
sns.set(rc={'figure.figsize': (27,7)})
ax = sns.countplot(x ='Product_Category', data =df )

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'])['Amount'].sum().reset_index()
```

```python
sales_state
```

```python
sns.set(rc={'figure.figsize': (27,7)})
sns.barplot(x ='Product_Category', y = 'Amount', data = sales_state)
```

# From above graphs we can see that most of sold products are clothing & Apparel, Food and Electronics gadgets

# Conclusion:

Married females age group 26-35 years from Karnataka working in IT are more likely to buy products from clothing Apparel, Food & Electronics Gadgets.
